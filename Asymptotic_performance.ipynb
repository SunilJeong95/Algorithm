```python
import matplotlib.pyplot as plt
import numpy as np
import math

N = np.arange(2, 20, 1)
log2 = [math.log2(n) if n>=2 else 0 for n in N]
nlog2 = [math.log2(n)*n if n>=2 else 0 for n in N]


plt.plot(N, N, 'r--', label="f(n)=n")
plt.plot(N, log2, 'g^', label="f(n)=log(n)")
plt.plot(N, nlog2, 'y*', label="f(n)=nlog(n)")
plt.plot(N, N**2, 'p', label="f(n)=n^2")
plt.plot(N, N**3, 'b', label="f(n)=n^3")
plt.plot(N, 2**N, 's', label="f(n)=2^n")
plt.ylim(0, 250)
plt.legend()
plt.show()
```
